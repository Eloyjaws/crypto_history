```python
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
doge = pd.read_csv('DOGE.csv')
```

```python
doge
```

```python
doge.info()
```

```python
fig, ax = plt.subplots()

ax.grid(True)

doge["date"] = pd.to_datetime(doge["date"])

ax.plot(doge.date, doge.close, color="red")
# ax.plot(x, new_y)
fig.suptitle(None)

ax.set_title("DOGECOIN'S PRICE HISTORY (2013 - 2021)")
ax.set_ylabel("Daily closing value (USD)")
ax.set_xlabel("Time")

fig.savefig("doge.png", dpi=300)
```

```python
doge2021 = doge[doge['date'].dt.year == 2021]

fig, ax = plt.subplots()

ax.grid(True)
ax.plot(doge2021.date, doge2021.close, color="red")
# ax.plot(x, new_y)
fig.suptitle(None)

ax.set_title("DOGECOIN'S PRICE HISTORY (2021)")
ax.set_ylabel("Daily closing value (USD)")
ax.set_xlabel("Time")

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

fig.savefig("doge2021.png", dpi=300)
```

```python
elon_influence_dates = pd.to_datetime(['2021-02-04', '2021-02-08'])
# elon_influence_dates = pd.to_datetime(['2020-12-2']) #  Another notable date is Dec 20, 2020

dogeFebruary = doge2021[doge2021['date'].dt.month == 2]

fig, ax = plt.subplots()

ax.grid(True)
value = ax.plot(dogeFebruary.date, dogeFebruary.close, color="red")
fig.suptitle(None)

ax.set_title("DOGECOIN'S PRICE HISTORY (FEB, 2021)")
ax.set_ylabel("Daily closing value (USD)")
ax.set_xlabel("Time")

for influence in elon_influence_dates:
    ax.vlines(influence, 0.02, 0.08, linestyles="-.", color="yellow")

ax.legend(['Coin Value', 'Day of Elon\'s viral influence'], loc=4)

for tick in ax.get_xticklabels():
    tick.set_rotation(45)

fig.savefig("doge_Feb_2021.png", dpi=300)
```
